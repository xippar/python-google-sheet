# ใช้ ezsheets เปิด Google Sheets

เราจะใช้ Library ที่ชื่อว่า ezsheets ที่ทำให้การเข้าถึง Google Sheets ทำได้ด้วยคำสั่งเพียงไม่กี่คำสั่ง

## ก่อนการใช้งาน

คำสั่งที่เราจะใช้งานเป็นคำสั่งที่อยู่ใน Library ezsheets เมื่อติดตั้งด้วยคำสั่ง !pip install ezsheets (หากยังไม่ติดตั้งให้ก็อปรันได้เลย) ให้ทำการ import มาก่อนเพื่อ ใช้งาน

```python
import ezsheets
```

## เปิด Spreadsheets ที่ต้องการ

### • ezsheet ทำงานกับ spreadsheet เท่านั้น

ในการทำงานกับ ezsheets จะทำงานร่วมกับ Spreadsheet เท่านั้น ซึ่งจะมีสัญลักษณ์เป็น icon เครื่องหมาย + แตกต่างจาก excel ที่มีสัญลักษณ์เป็นตัว x

### • สร้าง copy ของไฟล์ excel เป็น spreadsheet

เพื่อเตรียมข้อมูลให้อยู่ในรูปแบบที่ ezsheets สามารถทำงานด้วยได้

### • ใช้ spreadsheet ID เพื่อระบุ sheet ที่ต้องการ

การระบุ spreadsheet ID สามารถทำได้โดยดูจาก ลิ้งค์ของ spreadsheet ที่อยู่หลัง /d/ และก่อน /edit

### • ใช้ ezsheets.Spreadsheet() นำข้อมูลจาก spreadsheet

ใช้ spreadsheet ID ใส่เป็น input เป็น object ประเภท string คำสั่งจะสร้าง object ข้อมูลขึ้นมา ซึ่งเราจะได้ข้อมูลในรูปแแบของ spreadsheet

```python
spreadsheet = ezsheets.Spreadsheet('1XZBxD747u9fLJnJwGS5li7N6fx-X8doz0CQKtDOMstU')
spreadsheet
```

## เปิด sheet ที่ต้องการด้วย ชื่อชีท

การเปิด sheet ที่ต้องการจาก spreadsheet ใช้การระบุชื่อชีทเป็น object ประเภท string โดยระบุโดยใช้ [ ] วงเล็บเหลี่ยม ซึ่งเราจะได้ข้อมูลในรูปแบบอง sheet

```python
sheet = spreadsheet['North']
sheet
```

## ดู sheet ทั้งหมดที่มีอยู่ใน spreadsheet

เมื่อได้ข้อมูล spreadsheets มา เราสามารถใช้ .sheets เพื่อดูว่ามีชีทอะไรอยู่ใน spreadsheet ดังกล่าวบ้าง ซึ่งเราจะได้ข้อมูลออกมาในรูปของ Tuple

```python
spreadsheet.sheets
```

## ดึง sheet ที่ต้องการจาก sheet ทั้งหมด

ใช้การระบุตำแหน่งของ sheet ที่ต้องการผ่าน [ ] เพื่อดึงข้อมูลออกจาก Tuple เช่นดึง sheet แรกจากทั้งหมด ใช้เลข 0 แทนตำแหน่งแรก

```python
spreadsheet.sheets[0]
```

## เรียกดูชื่อชีทจากชีท

ใช้คำสั่ง .title ตามหลัง sheet เพื่อแสดงชื่อเป็น string

```python
spreadsheet.sheets[0].title
```

```python
sheet.title
```

## การดึงข้อมูลใน cell จาก sheet

เมื่อดึงข้อมูล sheet ใดๆ ได้แล้ว เราสามารถดึงข้อมูลจาก cell จาก sheet นั้นๆ ได้โดยการระบุ cell ที่ต้องการ และใช้สัญลักษณ์ [ ] วงเล็บเหลี่ยมเพื่อระบุตำแหน่งและดึงข้อมูลออกมา

### • การระบุ cell

หนึ่งในวิธีที่ตรงไปตรงมาที่สุดคือการเรียกตำแหน่ง Cell โดยตรง เช่น Column A แถวที่ 1 ใช้ 'A1'

```python
sheet['A1']
```

### • การระบุตำแหน่ง Column, Row

เป็นอีกวิธีในการระบุตำแหน่ง Cell ที่ต้องการด้วยการระบุตำแหน่งในลักษณะดังต่อไปนี้ Column, Row โดยตำแหน่งแรกเริ่มต้นที่ 1 ดังนั้นหากต้องการเรียก cell A1 จะใช้ 1,1

ข้อดีของการใช้การเรียกด้วยตัวเลขตำแหน่ง คือ เราสามารถใช้การใช้ loop ที่มีการเพิ่มจำนวนตัวเลขตาม loop เพื่อทำงานร่วมกับหลายข้อมูลได้ง่ายกว่า

```python
sheet[1,1]
```

## การแก้ไขข้อมูลใน sheet

การแก้ไขข้อมูลใน cell ใดๆ สามารถทำได้โดยการใช้ = เพื่อแก้ข้อมูลใน cell นั้นๆ ด้วยข้อมูลใหม่ได้ทันที

```python
sheet['A1'] = 'New Product'
sheet['A1']
```
