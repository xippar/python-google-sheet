# การดูข้อมูลทั้งหมดใน spreadsheets

มาดูวิธีการเขียนคำสั่งดึงข้อมูลทั้งหมดที่อยู่ใน spreadsheet กัน

## ก่อนการใช้งาน

คำสั่งที่เราจะใช้งานเป็นคำสั่งที่อยู่ใน Library ezsheets เมื่อติดตั้งด้วยคำสั่ง !pip install ezsheets (หากยังไม่ติดตั้งให้ก็อปรันได้เลย) ให้ทำการ import มาก่อนเพื่อ ใช้งาน

```python
import ezsheets
```

ก่อนการทำงานจะต้องดึงข้อมูล spreadsheet และ sheet มาด้วยก่อน

```python
spreadsheet = ezsheets.Spreadsheet('1XZBxD747u9fLJnJwGS5li7N6fx-X8doz0CQKtDOMstU')
sheet = spreadsheet['North']
sheet
```

## การดึง Column

ใช้คำสั่ง .getColumn() ตามหลังข้อมูล Sheet ซึ่งข้อมูลทั้งหมดใน Column จะถูกดึงออกมาในรูปแบบของ List

### • การระบุ Column ด้วยชื่อ

สามารถใช้ชื่อ Column ระบุ Column ที่ต้องการได้ โดยการระบุเช่น 'A' เป็น string เป็น input สำหรับ .getColumn('A') สังเกตตัวอักษรใหญ่เล็กในคำสั่ง

```python
sheet.getColumn('A')
```

### • การระบุ Column ด้วยตำแหน่ง

นอกจากการระบุชื่อ column เรายังสามารถระบุตำแหน่งของ column ที่ต้องการได้ เช่น column A คือตำแหน่งที่ 1 ดังนั้น สามารถใช้ 1 แทน 'A' ในคำสั่ง getColumn ได้เลย

```python
sheet.getColumn(1)
```

## การดึง Row

สามารถใช้ตำแหน่ง row ดึงข้อมูลทั้งหมดใน row นั้นๆ ออกมาเป็น List ได้ด้วยคำสั่ง .getRow() เช่น ดึง row แรกให้ใส่ .getRow(1)

```python
sheet.getRow(1)
```

## จัดการข้อมูลใน spreadsheet

สังเกตว่าเมื่อดึงข้อมูลแต่ละ column, row ออกมา จะมีการดึง '' หรือ string ที่ไม่มีข้อมูลออกมาเป็นจำนวนมาก เนื่องจากหากสำรวจที่ spreadsheet จะพบว่ามี cell จำนวนมากที่มีอยู่ แต่ไม่มีข้อมูลอยู่ในนั้น จึงเกิดเป็นการดึง object ที่ไม่มีข้อมูลอยู่ออกมานั่นเอง

ซึ่งการแก้ไขที่ต้องทำเพื่อให้จัดการข้อมูลได้ง่ายขึ้นนั่นคือ การกำจัดช่องว่างเหล่านี้ออกไปนั่นเอง

### ตัวอย่างข้อมูลจริง

สังเกตว่าจะมีช่องสีขาวที่ไม่มีข้อมูลอยู่ ซึ่งเป็นที่มาของ '' ที่ติดอยู่ใน List จำนวนมาก เมื่อใช้คำสั่ง .getColumn(), .getRow()

### การจัดการช่องว่าง

สามารถทำได้ผ่านการเข้าเปลี่ยนจำนวนของ Column, Row ที่มีอยู่ใน sheet เพื่อให้ spreadsheet เลือกเก็บข้อมูลแค่จำนวนที่กำหนดเท่านั้น ซึ่งจำนวนที่จะนำมาใช้ก็คือจำนวนข้อมูลที่มีอยู่ในแต่ละ Column, Row นั่นเอง และการที่จะได้มาซึ่งตัวเลขจำนวนนั้น อาจจะได้จากการนับ หรือการสร้างระบบนับข้อมูลที่มีอยู่ก็สามารถทำได้

### • จำนวน Column

สามารถดูจำนวน Column ปัจจุบันได้จากคำสั่ง .columnCount ซึ่งจะสังเกตว่ามีจำนวนมากกว่าจำนวนของมูลจริงที่มีอยู่เพียง 5 Column เท่านั้น

```python
sheet.columnCount
```

### • แก้ไขไจำนวน Column

สามารถแก้ไขโดยการใช้ = และตามด้วยจำนวนใหม่ที่ต้องการระบุได้ทันที เช่น ข้อมูลจริงมีอยู่เพียง 5 Column ก็ทำการระบุ = 5 ได้เลย

```python
sheet.columnCount = 5
sheet.columnCount
```

### • จำนวน Row

สามารถใช้คำสั่ง .rowCount เพื่อดูจำนวน row ในปัจจุบันซึ่งเกินไปจากข้อมูลจริงที่มีเพียง 12 row เท่านั้น

```python
sheet.rowCount
```

### • แก้ไขจำนวน Row

เราสามารถใช้คำสั่ง .rowCount และระบุ จำนวน row ที่มีอยู่เพื่อเปลี่ยนแปลงข้อมูลได้ในลักษณะเดียวกันกับการใช้เปลี่ยนค่า Column

```python
sheet.rowCount = 12
sheet.rowCount
```

### ข้อมูลหลังจัดการจำนวน Column, Row

หลังจากการจัดการจำนวน Column, Row สังเกตว่าการเปลี่ยนแปลงที่หน้า spreadsheet ข้อมูลที่ไม่มีอยู่ถูกลบออกไปเรียบร้อยแล้ว และเมื่อดึงข้อมูล column, row มาก็จะไม่มี '' ติดมาอีก

## การดึงข้อมูลหลาย Column

ใช้คำสั่ง .getColumns() ซึ่งจะรับ 2 ค่า คือ ตำแหน่งเริ่มต้นในการดึง และตำแหน่งหยุดดึงของ Column ซึ่งผลลัพธ์จะได้การดึงข้อมูลออกมาเป็นช่วงในรูปแบบของข้อมูลที่เป็น List ซ้อนอยู่ใน List

ข้อสังเกตคือ การดึงข้อมูลหลาย Column หรือการดึงเป็นช่วง จะดึงข้อมูลในตำแหน่งเริ่มดึงมา แต่จะไม่ดึงข้อมูลในตำแหน่งสุดท้ายมา ดังนั้นถ้าระบุการดึง แถวแรกที่ 1 และแถวสุดท้ายที่ 3 จะได้ แถวที่ 1,2 มารวม 2 แถว

```python
sheet.getColumns(1,3)
```

## การดึงข้อมูลหลาย Row

ใช้คำสั่ง .getRows() ซึ่งจะรับ 2 ค่า คือ ตำแหน่งเริ่มต้นในการดึง และตำแหน่งหยุดดึงของ Row ซึ่งผลลัพธ์จะได้การดึงข้อมูลออกมาเป็นช่วงในรูปแบบของข้อมูลที่เป็น List ซ้อนอยู่ใน List

เช่นเดียวกันกับ .getColumns() จะดึงตัวเริ่มต้น แต่จะไม่ดึงตัวสุดท้ายมา

```python
sheet.getRows(1,3)
```

## การแก้ไขข้อมูล / เพิ่มข้อมูล

การแก้ไขข้อมูล หรือ เพิ่มข้อมูลใน Column, Row นั้นใช้วิธีเดียวกัน แต่จะมีความแตกต่างที่การแก้ไขข้อมูลใช้การระบุ Column, Row ที่มีอยู่แล้วในลักษณะของการระบุข้อมูลเพื่อนำไปบันทึกทับข้อมูลเดิม แต่สำหรับการเพิ่มข้อมูลนั้นใช้การระบุ Column, Row ที่ยังไม่มีข้อมูลอยู่ ทำให้เกิดการสร้าง Column, Row ใหม่

### • การการแก้ไข / เพิ่ม Column

ใช้คำสั่ง .updateColumn() ที่จะรับข้อมูล 2 ค่า นั่นคือ ตำแหน่งของ Column ที่ต้องการเปลี่ยนแปลง / เพิ่มเข้าไป และข้อมูลที่ต้องการบันทึกลงไปใน Column นั้น

ตัวอย่าง การใส่ Column ใหม่ที่ได้มาจากการคำนวนผลรวมของทุก Column

```python
all_year = ['All Year',25940,18947,5032,2435,2368,2103,2330,2365,1664,1656,1942]
```

```python
sheet.updateColumn(6,all_year)
```

### • การการแก้ไข / เพิ่ม Row

ใช้คำสั่ง .updateRow() ที่จะรับข้อมูล 2 ค่า นั่นคือ ตำแหน่งของ Row ที่ต้องการเปลี่ยนแปลง / เพิ่มเข้าไป และข้อมูลที่ต้องการบันทึกลงไปใน Row นั้น

```python
all_product = ['All Product',14949,18943,15039,17851,66782]
```

```python
sheet.updateRow(13,all_product)
```
